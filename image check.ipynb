```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
DATA_DIR = "./data"
```

```python
for dir_ in os.listdir(DATA_DIR):
    dir_path = os.path.join(DATA_DIR, dir_)
    
    # Check if the current entry is a directory
    if os.path.isdir(dir_path):
        img_paths = os.listdir(dir_path)
        
        # Ensure there are image files in the directory
        if img_paths:
            first_img_path = img_paths[0]  # Get the path to the first image
            img = cv2.imread(os.path.join(dir_path, first_img_path))
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure()
plt.imshow(img_rgb)
plt.show()  # Add this line to display the image
```
